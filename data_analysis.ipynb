```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./Data/forms_cleaned.csv')
df.head()
```

```python
df.info()
```

```python
# show in a graph the column 3 of data frame binary
sns.set_style('whitegrid')
fig = sns.countplot(x='¿Ha asistido a estudios de grabación y/o de producción musical?', data=df)

save_figure =  fig.get_figure()
save_figure.savefig('./Images/Question3.png')
```

```python
# Data
labels = 'No', 'Sí'
sizes = df['¿Crees que hay suficientes oportunidades para los estudiantes de música para conectarse con otros músicos y colaborar en proyectos?'].value_counts()

print(sizes)

# Plot
fig, ax = plt.subplots()

ax.set_title('¿Crees que hay suficientes oportunidades para los estudiantes de música para conectarse con otros músicos y colaborar en proyectos?')
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Show plot
plt.show()
save_figure =  fig.get_figure()
save_figure.savefig('./Images/piechart1.png')
```

```python
# Data
labels = ['Sí']
sizes = df['¿Piensas que los estudios de grabación y/o de producción musical da espacio para conocer a personas que puedan ayudar a tu crecimiento profesional?'].value_counts()

print(sizes)

# Plot
fig, ax = plt.subplots()

ax.set_title('¿Piensas que los estudios de grabación y/o de producción musical da espacio para conocer a personas que puedan ayudar a tu crecimiento profesional?')
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Show plot
plt.show()
save_figure =  fig.get_figure()
save_figure.savefig('./Images/piechart2.png')
```

```python
# pass to an array '¿Qué tipo de recursos utilizas para estudiar música?'?
values = df['¿Qué tipo de recursos utilizas para estudiar música?'].value_counts()
print(values)
```

```python
# videos, Paginas web, Libros, Otros
videos_sum = values[0] + values[1] + values[2] + values[3] + values[5]
web_sum = values[0] + values[3] + values[4] + values[5]
books_sum = values[2] + values[3]
others_sum = values[5]

values_sum = [videos_sum, web_sum, books_sum, others_sum]
print(values_sum)
```

```python
#histogram with values_sum
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title('¿Qué tipo de recursos utilizas para estudiar música?')
ax.set_ylabel('Frecuencia')
ax.set_xlabel('Recursos')
ax.bar(['Videos', 'Paginas web', 'Libros', 'Otros'], values_sum)
plt.show()
# save the figure
save_figure =  fig.get_figure()
save_figure.savefig('./Images/Question6.png')
```

```python
# pass to an array '¿Por qué no asistes a estudios de grabación y/o de producción musical?'
values = df['¿Por qué no asistes a estudios de grabación y/o de producción musical?'].value_counts()
print(values)
```

```python
# Por el costo, Por la poca adecuacion del espacio, La ubicación del sitio, Otro
cost_sum = values[1] + values[2] + values[4]
space_sum = values[0] + values[4]
location_sum = values[1]
other_sum = values[3] + values[5]

values_sum = [cost_sum, space_sum, location_sum, other_sum]
print(values_sum)
```

```python
# histogram with values
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title('¿Por qué no asistes a estudios de grabación y/o de producción musical?')
ax.set_ylabel('Frecuencia')
ax.set_xlabel('Razones')
ax.bar(['costo', 'poca adecuacion', 'ubicación', 'Otro'], values_sum)
plt.show()

# save the figure
save_figure =  fig.get_figure()
save_figure.savefig('./Images/Question8.png')
```
